```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_dropped = housing.drop(columns=['url','id','Cid','Unnamed: 0'])
housing_dropped
```

```python
housing_null=pd.DataFrame(housing_dropped.isna().sum())
housing_null.head()
```

```python
housing_dropped.DOM.plot(kind = 'box')
```

```python
housing_no_missing = housing_dropped.copy()
```

```python
housing_no_missing['DOM'].fillna(float(housing_no_missing['DOM'].dropna().mode()),inplace=True)
housing_no_missing.dropna(subset=['subway','elevator'],how='any',inplace=True)
housing_no_missing
```

```python
housing_no_missing.info()
```

```python
housing_no_missing.to_csv('housing_no_missing.csv', encoding='gbk', index=False)
housing_no_missing
```

```python
housing_categorical=housing_no_missing.copy()
housing_categorical["elevator"] = housing_categorical["elevator"].map({1: "has elevator",0:"no elevator"})
housing_categorical['subway'] = housing_categorical["subway"].map({1: "has subway",0:"no subway"})
housing_categorical['buildingStructure'] = housing_categorical['buildingStructure'].map({1:'unknown',2:'mixed',3:'brick and wood',4:'concrete',5: 'steel',6:'steel-concrete composite'})
housing_categorical['renovationCondition'] =housing_categorical['renovationCondition'].map({1:'other',2:'rough',3:'Simplicity',4:'hardcover'})
housing_categorical.head(10)
```

```python
print('constructionTime : ',housing_categorical.constructionTime.unique())
print('floor : ',housing_categorical.floor.unique())
```

```python
housing_construction = housing_categorical.copy()
housing_construction.loc[housing_construction['constructionTime'].str.contains(r'[\u4e00-\u9fff]'), 'constructionTime'] = np.nan
housing_construction = housing_construction.dropna(subset=['constructionTime'])
housing_construction['constructionTime'] = housing_construction['constructionTime'].astype(int)
housing_construction['constructionTime'].fillna(housing_construction['constructionTime'].mean(), inplace=True)
housing_construction['constructionTime'] = housing_construction['constructionTime'].astype('int32')
housing_construction.head()
```

```python
housing_floor = housing_construction.copy()
housing_floor['floor'] = housing_floor['floor'].str.replace(r'\D','', regex=True)
```

```python
housing_floor.totalPrice.plot(kind = 'box')
```

```python
Q1 = housing_floor['totalPrice'].quantile(0.25)
Q3 = housing_floor['totalPrice'].quantile(0.75)
IQR = Q3 - Q1
housing_no_outlier = housing_floor.drop(housing_floor[ (housing_floor.totalPrice < Q1-1.5*IQR) | (housing_floor.totalPrice> Q3+1.5*IQR)].index)
print('Number of removed outliers : ', housing_floor.shape[0] - housing_no_outlier.shape[0])
```

```python
housing_no_outlier.totalPrice.plot(kind = 'box')
```

```python
housing_no_outlier.to_csv('housing_no_outlier.csv', encoding='gbk', index=False)
```

```python
from math import radians
capital_Lng = radians(116.4074)
capital_Lat = radians(39.9042)
housing_capital = housing_no_outlier.copy()
Lat = housing_capital['Lat'].apply(lambda x: radians(x))
Lng = housing_capital['Lng'].apply(lambda x: radians(x))
x =  np.arccos(np.sin(Lat)*np.sin(capital_Lat) + np.cos(Lat)*np.cos(capital_Lat)*np.cos(capital_Lng-Lng))
housing_capital['distanceToCapital'] = x*6371.0088
housing_capital.head()
```

```python
housing_PPS = housing_capital.copy()

housing_PPS['pricePerSquare'] = (housing_PPS.totalPrice/housing_PPS.square)*1000

housing_PPS.head()
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(15,15)
housing_PPS.hist(ax = ax, bins = 20);
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(8,8)

sns.regplot(ax = ax, data = housing_PPS, x = 'distanceToCapital', y = 'pricePerSquare', 
                   line_kws={'color':'red'}, scatter_kws= {'alpha': 0.1});
```

```python
fig, ax =plt.subplots(1,1,figsize=(6,6))
a=housing_PPS[housing_PPS['elevator']=='has elevator']
b=housing_PPS[housing_PPS['elevator']=='no elevator']
b.totalPrice.plot(kind='kde',ax=ax,label='no elevator')
a.totalPrice.plot(kind='kde',ax=ax,label='has elevator')
ax.set_xlabel('pricePerSquare')
ax.set_ylabel('Density')
ax.set_title('effect of elevator on price',color='red')
ax.legend()
ax.set_facecolor('#ffffcc')
fig.set_facecolor('khaki')
```

```python
housing_PPS.to_csv('housing_extended.csv', encoding='gbk', index=False)
```

```python
housing_extended = pd.read_csv('housing_extended.csv', encoding='gbk')
```

```python
districts_df = pd.DataFrame()
districts_df['pricePerSquare mean'] = housing_extended.groupby('district')['pricePerSquare'].mean()
districts_df['% houses with elevator'] = housing_extended[housing_extended.elevator=='has elevator'].groupby('district').size()/housing_extended.groupby(['district']).size()*100
districts_df['% houses with subway'] = housing_extended[housing_extended.subway=='has subway'].groupby('district').size()/housing_extended.groupby(['district']).size()*100
districts_df['square mean'] = housing_extended.groupby('district')['square'].mean()
districts_df['building age mean'] = 2021-housing_extended.groupby('district')['constructionTime'].mean()
districts_df['frequent buildingStructure'] = housing_extended.groupby(['district'])['buildingStructure'].agg(lambda x:x.value_counts().idxmax())
districts_df
```

```python
housing_extended['tradeTime'] = pd.to_datetime(housing_extended['tradeTime'])
housing_extended.head()
```

```python
fig1 , ax1 = plt.subplots(figsize=(8,8))
month3 = housing_extended[housing_extended.tradeTime>='2010'].resample(rule='M',on='tradeTime')['pricePerSquare'].mean().plot(ax=ax1,kind='line',color='crimson')
ax1.set_title('pricePerSquare over time')
ax1.set_ylabel('pricePerSquare')
```

```python
fig2 , ax2 = plt.subplots(figsize=(8,8))
month_3 = housing_extended[(housing_extended.tradeTime>'2010')].resample(rule='3M',on='tradeTime')['distanceToCapital'].apply(lambda x:(x<15).sum())
month_3.plot(kind='line',color='orange',ax=ax2,legend=False)
ax2.set_ylabel('frequency')
ax2.set_title('center city house trades over time')
```
